#  Data manipulation: sales dataset + visualization 

```python
## read csv data

sales_data = pd.read_csv('/kaggle/input/sales-data/data.csv')
print(sales_data.shape)
sales_data.head()
```

```python
# check the datatypes
sales_data.dtypes
```

```python
# check the missing values
sales_data.isnull().sum()
```

```python
# convert Date into datetime datatype
sales_data['Date'] = pd.to_datetime(sales_data['Date'])
```

```python
sales_data.dtypes
```

```python
sales_data['Date'].dt.year.head()
```

```python
# select only 'sales, Category' columns
sales_data[['Sales','Category']].head()
```

```python
# filter rows where Region is 'North'
sales_data[sales_data['Region'] == 'North']
```

```python
# select top sales i.e sales > 800

sales_data[sales_data['Sales'] > 800].head()
```

```python
# sort the dataset by date in descending order

sorted_df = sales_data.sort_values(by='Date', ascending=False)
sorted_df.head()
```

```python
print(sales_data['Sales'].mean())
```

```python
# fill missing data
sales_data['Sales'] = sales_data['Sales'].fillna(sales_data['Sales'].mean())
sales_data['Value'] = sales_data['Value'].fillna(sales_data['Value'].mean())
sales_data.sort_values(by='Date', ascending=False).head()
```

# Grouping and Aggregation

```python
# grouping and aggregation
sales_data.groupby('Region')['Value'].mean()
```

```python
sales_data.groupby('Category')['Sales'].max()
```

```python
sales_data.groupby('Category')['Sales'].sum()
```

```python
sales_data.groupby('Product')['Sales'].sum()
```

```python
# multi-aggregation
sales_data.groupby('Product')['Sales'].agg(['sum','mean','count']).rename(columns={"sum": "Total_Sales", "mean": "Average_Sales", "count": "Count_Sales"})
```

```python
# group by multiple column
print(sales_data.groupby(['Sales','Category']).first().head())
```

```python
# data Transformation, percentage contribution of each productâ€™s Sales
sales_data['Sales_pct_of_category'] = sales_data.groupby('Category')['Sales'].transform(lambda x: x / x.sum() * 100)
sales_data.head()
```

```python
#  how much each Sale differs from the Category average?
sales_data['Sales_diff_from_category_avg'] = sales_data['Sales'] - sales_data.groupby('Category')['Sales'].transform('mean')
sales_data.head(5)
```

```python
# Filter categories with total Sales > 150000
filtered_df = sales_data.groupby('Category').filter(lambda x: x['Sales'].sum() > 9000)

filtered_df.head(5)
```

```python
# filter regions with mean value over 50
filtered_region_over_100 = sales_data.groupby('Region').filter(lambda x: x['Value'].mean() >50)
filtered_region_over_100.head()
```

# Date Handling

```python
# total monthly sales
sales_data['Year'] = sales_data['Date'].dt.year
sales_data['Month'] = sales_data['Date'].dt.month
sales_data['Day'] = sales_data['Date'].dt.day

# use name of the month
sales_data['Month_Name'] = sales_data['Date'].dt.strftime('%B')
sales_data.head()
```

```python
# Total monthly sales
monthly_sales = sales_data.groupby(['Year', 'Month_Name'])['Sales'].sum().reset_index()
monthly_sales

# if this Date was the index
# df = df.set_index('Date')
# monthly_sales = df['Sales'].resample('M').sum()
# print(monthly_sales)
```

```python
# maximum monthly value
monthly_value = sales_data.groupby(['Year', 'Month'])['Value'].sum().reset_index()
max_month = monthly_value.loc[monthly_value['Value'].idxmax()]
max_month
```
